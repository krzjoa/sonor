```python
import numpy as np
import scipy
import IPython
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
# Parametry nagrania
sr = 22050 # sample rate
T = 5.0    # seconds
```

```python
t = np.linspace(0, T, int(T*sr), endpoint=False)
s1 = 0.5 * np.sin(2*np.pi*440*t)
s2 = 0.5 * np.sin(2*np.pi*554.365*t)
s3 = 0.5 * np.sin(2*np.pi*659.255*t)
s4 = 0.5 * np.sin(2*np.pi*783.991*t)

s = s1 + s2 + s3 + s4
```

```python
plt.figure(figsize=(20,10))
plt.plot(t[1:2000], s[1:2000])
plt.show()
```

```python
ipd.Audio(s, rate=sr)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import (
    Audio, display, clear_output)
from ipywidgets import widgets
from functools import partial
%matplotlib inline
```

```python
rate = 16000.
duration = .25
t = np.linspace(
    0., duration, int(rate * duration))
```

```python
def synth(f):
    x = np.sin(f * 2. * np.pi * t)
    display(Audio(x, rate=rate, autoplay=True))
```

```python
synth(600)
```

```python
notes = 'C,C#,D,D#,E,F,F#,G,G#,A,A#,B,C'.split(',')
freqs = 440. * 2**(np.arange(3, 3 + len(notes)) / 12.)
notes = list(zip(notes, freqs))
```

```python
layout = widgets.Layout(
    width='30px', height='60px',
    border='1px solid black')

buttons = []
for note, f in notes:
    button = widgets.Button(
        description=note, layout=layout)

    def on_button_clicked(f, b):
        # When a button is clicked, we play the sound
        # in a dedicated Output widget.
        with widgets.Output():
            synth(f)
            print("hohoo")

    button.on_click(partial(on_button_clicked, f))
    buttons.append(button)

# We place all buttons horizontally.
widgets.Box(children=buttons)
```
